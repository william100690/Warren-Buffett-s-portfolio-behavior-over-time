```python
import pandas as pd
import numpy as np
import openpyxl
```

```python
#read the file from its relative path

buffet_2022 = pd.read_excel(r'Data_sets\Buffet_2022Q2.xlsx')
buffet_2022.head(3)
```

Number of companies in Berkshire Hathaway's portfolio by year and total worth

```python
# replace all the 0.0 and 0 values with NaN
buffet_2022 = buffet_2022.replace(0.0, np.nan)
# change the 1st column name to year
buffet_2022 = buffet_2022.rename(columns={'Unnamed: 0':'Year'})
buffet_2022.head(3)
```

```python
#clean the first column by leaving just the year and make it a int variable
# create a function to clean the column
def clean_column(word):
    word = str(word)
    word = word.replace("-12-31 00:00:00","")
    word = word.replace("-06-30 00:00:00","")
    word = int(word)
    return word
```

```python
#Apply the function to the column
buffet_2022["Year"] = buffet_2022["Year"].apply(lambda x: clean_column(x))
buffet_2022.head(3)
```

```python
#create a new column called "Total/companies" and set the number of companies
buffet_2022["Total/companies"] = buffet_2022.count(axis=1)

#create a new column called "Total/worth" and set the worth of the porfolio
buffet_2022["Total/worth"] = buffet_2022.sum(axis=1)

#make a new data frame with the Total/companies and Total/woth per year
buffet_compnies_worht = buffet_2022[["Year","Total/companies","Total/worth"]]
buffet_compnies_worht.head(3)
```

```python
#Convert the data frame into an Excel file to perform a better visualization with the Tableau tool.
buffet_compnies_worht.to_excel('Data_sets\Buffet_companies_worth.xlsx', index=False)
```
